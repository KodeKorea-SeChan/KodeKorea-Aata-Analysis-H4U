```python
import requests
import json
import time
import pandas as pd
from tqdm import tqdm

api_key = 'RGAPI-313f4853-dfb4-4f5e-982e-ca429ccf760a'
```

```python
#챌린저 등급 솔로 랭크 게임을 플레이한 플레이어의 summoner id 추출
url = 'https://kr.api.riotgames.com/lol/league/v4/challengerleagues/by-queue/RANKED_SOLO_5x5?api_key=' + api_key

summonerId = {}

r = requests.get(url)
r = r.json()['entries']

num = 0

for i in r:
    summonerId[i['summonerName']] = i['summonerId']
    num += 1
print(num)
```

```python
from numpy.lib.function_base import kaiser
# summonerId를 통해 puuid 추출
print(summonerId)

allcnt = 0
bluewincnt = 0

for i,j in zip(tqdm(summonerId.values()),summonerId.keys()):
  try:
    # summoner id를 가지고 puuid 조회
    url2 = 'https://kr.api.riotgames.com/lol/summoner/v4/summoners/' + i + '?api_key=' + api_key
    r = requests.get(url2)
    r.json()

    # puuid를 가지고 matchid 조회
    url_match = 'https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/'+ r.json()['puuid'] +'/ids?start=1&count=5&api_key='+api_key
    r_match = requests.get(url_match)

    # matchid를 가지고 게임 내부기록 조회
    url_match = 'https://asia.api.riotgames.com/lol/match/v5/matches/' + r_match.json()[0] + '?api_key=' + api_key
    r_match = requests.get(url_match)

    # 블루진영 레드진영 이깃횟수 카운팅
    for k in range(len(r_match.json()['info']['participants'])):
      if r_match.json()['info']['participants'][k]['summonerName'] == j:
          if r_match.json()['info']['participants'][k]['teamId'] == 100:
              bluewincnt+=1
              allcnt+=1
          else:
              allcnt+=1
  except:
    continue

print()
print('블루팀이 이긴횟수: ', bluewincnt/allcnt)
```
